```python
weather = pd.read_csv('./data-4/weather.csv')
weather
```

```python
weather_melt = pd.melt(weather, id_vars=['id', 'year', 'month', 'element'], 
                       var_name='day',
                       value_name='temp')
```

```python
weather_melt
```

```python
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp',
    dropna=False
)

##index 인자에는 위치를 그대로 유지할 열 이름을 지정하고,
#columns 인자에는 피벗할 열 이름을 지정하고,
#values 인자에는 새로운 열의 데이터가 될 열 이름을 지정
```

```python
weather_tidy
```
